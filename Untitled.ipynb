```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots show inline in the notebook
%matplotlib inline

print("Libraries loaded ✓")
```

```python
np.random.seed(42)  # makes results reproducible — same "random" data every time
n = 2000            # 2000 synthetic customers

df = pd.DataFrame({
    "customer_id":           [f"CUST-{i:04d}" for i in range(n)],
    "days_since_last_order": np.random.exponential(45, n).astype(int).clip(1, 365),
    "total_orders":          np.random.poisson(6, n).clip(1, 50),
    "avg_order_value":       np.random.lognormal(4.2, 0.7, n).round(2),
    "support_tickets":       np.random.poisson(0.8, n).clip(0, 10),
    "returns_count":         np.random.poisson(0.4, n).clip(0, 8),
    "email_open_rate":       np.random.beta(2, 5, n).round(3),
    "days_since_signup":     np.random.randint(30, 1200, n),
    "promo_used_count":      np.random.poisson(2, n).clip(0, 15),
    "country": np.random.choice(
        ["US", "UK", "CA", "AU", "DE", "FR", "Other"], n,
        p=[0.35, 0.18, 0.12, 0.08, 0.07, 0.07, 0.13]
    ),
    "acquisition_channel": np.random.choice(
        ["Organic", "Paid Social", "Email", "Referral", "Influencer"], n,
        p=[0.30, 0.25, 0.20, 0.15, 0.10]
    ),
})

# Total spend = orders × average order value
df["total_spend"] = (df["total_orders"] * df["avg_order_value"]).round(2)

print(f"Dataset shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
```

```python
#churn score
# Churn score — weighted combination of risk signals
churn_score = (
    0.25 * (df["days_since_last_order"] / 365) +  # long inactivity = higher risk
    0.20 * (1 - df["email_open_rate"]) +           # not opening emails = disengaged
    0.15 * (df["support_tickets"] / 10) +          # lots of complaints = at risk
    0.15 * (df["returns_count"] / 8) +             # lots of returns = unhappy
    0.10 * (1 / (df["total_orders"] + 1)) +        # few orders = low loyalty
    0.15 * np.random.uniform(0, 1, n)              # some randomness — real life is messy
)

# Anyone above 0.42 threshold is labelled churned
df["churned"] = (churn_score > 0.42).astype(int)

# What's our churn rate?
churn_rate = df["churned"].mean() * 100
print(f"Overall churn rate: {churn_rate:.1f}%")
print(f"Churned customers: {df['churned'].sum()}")
print(f"Retained customers: {(df['churned'] == 0).sum()}")
```

```python
#visualize 
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Churn signal exploration", fontsize=14, fontweight='bold')

# 1. Days since last order — churned vs retained
axes[0,0].hist(df[df["churned"]==0]["days_since_last_order"], 
               bins=40, alpha=0.6, label="Retained", color="steelblue")
axes[0,0].hist(df[df["churned"]==1]["days_since_last_order"], 
               bins=40, alpha=0.6, label="Churned", color="tomato")
axes[0,0].set_title("Days since last order")
axes[0,0].legend()

# 2. Email open rate
axes[0,1].hist(df[df["churned"]==0]["email_open_rate"], 
               bins=40, alpha=0.6, label="Retained", color="steelblue")
axes[0,1].hist(df[df["churned"]==1]["email_open_rate"], 
               bins=40, alpha=0.6, label="Churned", color="tomato")
axes[0,1].set_title("Email open rate")
axes[0,1].legend()

# 3. Churn rate by acquisition channel
channel_churn = df.groupby("acquisition_channel")["churned"].mean().sort_values(ascending=False)
axes[1,0].bar(channel_churn.index, channel_churn.values, color="coral")
axes[1,0].set_title("Churn rate by channel")
axes[1,0].tick_params(axis='x', rotation=30)

# 4. Total spend vs churn
axes[1,1].boxplot(
    [df[df["churned"]==0]["total_spend"], df[df["churned"]==1]["total_spend"]],
    labels=["Retained", "Churned"]
)
axes[1,1].set_title("Total spend distribution")
axes[1,1].set_ylabel("$")

plt.tight_layout()
plt.show()
```

BUILDING THE ML MODEL

```python
from sklearn.preprocessing import LabelEncoder

# These are the columns we'll feed into the model
# We exclude customer_id (just a label) and churned (that's what we're predicting)
features = [
    "days_since_last_order",
    "total_orders",
    "avg_order_value",
    "total_spend",
    "support_tickets",
    "returns_count",
    "email_open_rate",
    "days_since_signup",
    "promo_used_count"
]

# ML models only understand numbers — not strings like "US" or "Organic"
# LabelEncoder converts text categories to integers
le_country = LabelEncoder()
le_channel = LabelEncoder()

df["country_enc"]  = le_country.fit_transform(df["country"])
df["channel_enc"]  = le_channel.fit_transform(df["acquisition_channel"])

# Add the encoded columns to our feature list
features += ["country_enc", "channel_enc"]

# X = inputs (features), y = output (what we want to predict)
X = df[features]
y = df["churned"]

print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")
print(f"\nFeature columns:\n{features}")


```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,       # 20% held back for testing
    random_state=42,     # reproducible split every time
    stratify=y           # keeps churn ratio equal in both splits
)

print(f"Training set:  {X_train.shape[0]} customers")
print(f"Test set:      {X_test.shape[0]} customers")
print(f"\nChurn rate in training: {y_train.mean()*100:.1f}%")
print(f"Churn rate in test:     {y_test.mean()*100:.1f}%")
```

```python
from sklearn.ensemble import GradientBoostingClassifier

model = GradientBoostingClassifier(
    n_estimators=120,    # number of decision trees to build
    max_depth=4,         # how deep each tree can go
    learning_rate=0.08,  # how much each tree corrects the previous one
    random_state=42
)

model.fit(X_train, y_train)

print("Model trained ✓")
print(f"\nTraining accuracy: {model.score(X_train, y_train)*100:.1f}%")
print(f"Test accuracy:     {model.score(X_test, y_test)*100:.1f}%")
```

```python
from sklearn.metrics import roc_auc_score, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Get probability scores (not just 0/1 predictions)
y_pred_proba = model.predict_proba(X_test)[:, 1]  # probability of churning
y_pred       = model.predict(X_test)              # hard 0/1 prediction

# AUC-ROC score
auc = roc_auc_score(y_test, y_pred_proba)
print(f"AUC-ROC Score: {auc:.3f}")
print()

# Full classification report
print(classification_report(y_test, y_pred, target_names=["Retained", "Churned"]))
```

```python
# Feature importance — what does the model rely on most?
importances = pd.Series(
    model.feature_importances_,
    index=features
).sort_values(ascending=True)

plt.figure(figsize=(10, 6))
importances.plot(kind="barh", color="steelblue")
plt.title("What drives churn? (feature importance)", fontweight="bold")
plt.xlabel("Importance score")
plt.tight_layout()
plt.show()

print("\nTop 3 churn drivers:")
for feat, score in importances.sort_values(ascending=False).head(3).items():
    print(f"  {feat}: {score:.3f}")
```

```python
# Predict probability for ALL customers (not just test set)
df["churn_probability"] = model.predict_proba(X[features])[:, 1]

# Bucket into risk tiers
df["risk_tier"] = pd.cut(
    df["churn_probability"],
    bins=[0, 0.35, 0.65, 1.0],
    labels=["Low Risk", "Medium Risk", "High Risk"]
)

# Quick summary
print(df["risk_tier"].value_counts())
print()
print("Top 5 highest risk customers:")
df.sort_values("churn_probability", ascending=False)[
    ["customer_id", "churn_probability", "days_since_last_order", "email_open_rate", "risk_tier"]
].head()
```
